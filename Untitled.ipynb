## Preparação de dados para modelos de ML

```python
import sklearn
```

```python
df_cleaned = df_rent.drop(["New", "Property Type", "Negotiation Type"], axis=1)
```

```python
df_cleaned
```

### Tratamento e dados categóricos

#### Jeito 1 de tratar dados dos distritos

```python
from sklearn.preprocessing import OrdinalEncoder

ordinal_encoder = OrdinalEncoder()
```

```python
district_encoded = ordinal_encoder.fit_transform(df_rent[["District"]])
```

```python
district_encoded
```

#### Jeito 2

```python
from sklearn.preprocessing import OneHotEncoder

cat_encoder = OneHotEncoder()
```

```python
housing_cat_1hot = cat_encoder.fit_transform(df_cleaned[['District']])
housing_cat_1hot
```

```python
housing_cat_1hot.toarray()
```

```python
one_hot = pd.get_dummies(df_cleaned["District"])
df = df_cleaned.drop('District', axis = 1)
```

```python
df = df.join(one_hot)
```

```python
df
```

#### Separando os dados de treino

```python
from sklearn.model_selection import train_test_split
```

```python
Y = df["Price"]
X = df.loc[:, df.columns != "Price"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3) ## 70% treino e 30% teste
```

```python
x_train
```

#### 5059 dados de treino

```python
# LinearRegression

from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(x_train, y_train) # Será feito a otimização do modelo utilizando a função custo
```

```python
alguns_dados = x_train.iloc[:5]
algumas_label = y_train.iloc[:5]

print(f"Predições: {lin_reg.predict(alguns_dados)}")
print(f"Labels: {algumas_label.values}")
```

#### Previu que valia 1972 e vale 2500 e assim sucessivamente 

```python
alguns_dados
```

```python
from sklearn.metrics import mean_squared_error
preds = lin_reg.predict(x_train)
lin_mse = mean_squared_error(y_train, preds)

lin_rmse = np.sqrt(lin_mse)
lin_rmse
```
